```python
import folium
import pandas as pd

# Sample data simulating your full CSV dataset
data = [
    {"subdivision": "Almatti Reservoir", "storage": 120.5, "live_capacity_frl": 100.0},
    {"subdivision": "Bargi Reservoir", "storage": 95.0, "live_capacity_frl": 90.0},
    {"subdivision": "Hirakud Reservoir", "storage": 110.0, "live_capacity_frl": 100.0},
    {"subdivision": "Tehri Reservoir", "storage": 80.0, "live_capacity_frl": 120.0},
    {"subdivision": "Bhatsa Reservoir", "storage": 160.0, "live_capacity_frl": 100.0},
    {"subdivision": "Sardar Sarovar", "storage": 99.0, "live_capacity_frl": 100.0},
    {"subdivision": "Upper Vaitarna Reservoir", "storage": 105.0, "live_capacity_frl": 100.0},
    {"subdivision": "Indira Sagar", "storage": 102.0, "live_capacity_frl": 100.0},
    {"subdivision": "Nagarjuna Sagar", "storage": 89.0, "live_capacity_frl": 100.0},
    {"subdivision": "Idukki Reservoir", "storage": 70.0, "live_capacity_frl": 100.0},
    {"subdivision": "Krishnaraja Sagar", "storage": 101.0, "live_capacity_frl": 100.0},
    {"subdivision": "Ramganga Reservoir", "storage": 130.0, "live_capacity_frl": 100.0},
]

# Coordinates for each subdivision
reservoir_locations = {
    "Almatti Reservoir": (16.3386, 75.8864),
    "Bargi Reservoir": (22.9933, 79.8897),
    "Bhatsa Reservoir": (19.6944, 73.3217),
    "Hirakud Reservoir": (21.5456, 83.8724),
    "Sardar Sarovar": (21.8228, 73.7191),
    "Tehri Reservoir": (30.3831, 78.4800),
    "Upper Vaitarna Reservoir": (19.85, 73.52),
    "Indira Sagar": (22.2514, 76.4741),
    "Nagarjuna Sagar": (16.5733, 79.3186),
    "Idukki Reservoir": (9.8435, 76.9784),
    "Krishnaraja Sagar": (12.4220, 76.5717),
    "Ramganga Reservoir": (29.5964, 78.7530),
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate flood risk
df["flood_risk"] = df["storage"] / df["live_capacity_frl"]

# Create a map centered in India
flood_map = folium.Map(location=[22.0, 78.0], zoom_start=5)

# Add markers for all reservoirs
for _, row in df.iterrows():
    name = row['subdivision']
    risk = row['flood_risk']
    location = reservoir_locations.get(name, (22.0, 78.0))  # Default center if missing

    # Color by risk level
    if risk > 1.5:
        color = "red"
    elif risk > 1.0:
        color = "orange"
    else:
        color = "green"

    folium.CircleMarker(
        location=location,
        radius=8,
        popup=f"{name}<br>Flood Risk: {risk:.2f}",
        color=color,
        fill=True,
        fill_color=color
    ).add_to(flood_map)

# Display map (in Jupyter or similar)
flood_map
```
